## Step 4: Data Cleaning

Fill missing with default values

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['cast'].fillna('Not Specified', inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
```

```python
pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['rating'].fillna("Unknown", inplace=True)
```

# Step 5: Content Type distribution

```python
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Content Type Distribution')
plt.show()
```

# Step 6: Country-wise Content

```python
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Top 10 Countries by Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

# Step 7: Rating Distribution

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index[:10], palette="Paired")
plt.title('Top Ratings on Netflix')
plt.show()
```

# Step 8: Genre Analysis

```python
df['listed_in'] = df['listed_in'].astype(str)
genres = df['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='mako')
plt.title("Top 10 Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```


# Step 9: Yearly Trend

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='year_added', hue='type', palette='coolwarm')
plt.title("Yearly Trend of Netflix Additions")
plt.xticks(rotation=45)
plt.show()
```
